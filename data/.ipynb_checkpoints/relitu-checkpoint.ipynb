# 热力图以及相关系数
## 静态数据一般采用此方法，时序性数据一般不采用

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore") #过滤掉警告的意思
from pyforest import *
```

```python

data = pd.read_csv('d1.csv', on_bad_lines='skip')
data.head()
```

```python
#图片显示中文
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] =False #减号unicode编码
data.drop(['Year','Month','Day','Hour','Minute'], axis=1, inplace=True) #删除无关的列
```

```python
data.head()
```

```python
#计算各变量之间的相关系数
corr = data.corr()
corr
```

```python
ax = plt.subplots(figsize=(20, 16))#调整画布大小
ax = sns.heatmap(corr, vmax=.8, square=True, annot=True)#画热力图   annot=True 表示显示系数
# 设置刻度字体大小
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```

# 皮尔逊相关

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv('d1.csv', on_bad_lines='skip')
df.drop(['Year','Month','Day','Hour','Minute'], axis=1, inplace=True) #删除无关的列

overall_pearson_r = df.corr().iloc[0,1]
# 使用Pandas计算数据框中第一列和第二列之间的皮尔逊相关系数
# 并将结果存储在overall_pearson_r中，然后将其打印出来。
print(f"Pandas computed Pearson r: {overall_pearson_r}")
```

```python
overall_pearson_r = df.corr().iloc[0,2]
print(f"Pandas computed Pearson r: {overall_pearson_r}")
overall_pearson_r = df.corr().iloc[0,3]
print(f"Pandas computed Pearson r: {overall_pearson_r}")
overall_pearson_r = df.corr().iloc[0,4]
print(f"Pandas computed Pearson r: {overall_pearson_r}")
overall_pearson_r = df.corr().iloc[0,5]
print(f"Pandas computed Pearson r: {overall_pearson_r}")

```

```python
df = pd.read_csv('d1.csv', on_bad_lines='skip')
df.drop(['Year','Month','Day','Hour','Minute'], axis=1, inplace=True) #删除无关的列

overall_pearson_r = df.corr().iloc[0,2]
print(f"Pandas computed Pearson r: {overall_pearson_r}")
# 计算滑动窗口同步性
f,ax=plt.subplots(figsize=(7,3))
df.rolling(window=30,center=True).median().plot(ax=ax)
ax.set(xlabel='Time',ylabel='Pearson r')
ax.set(title=f"Overall Pearson r = {np.round(overall_pearson_r,2)}")
plt.show()
```

```python
# 计算滑动窗口同步性
f,ax=plt.subplots(figsize=(7,3))
df.rolling(window=30,center=True).median().plot(ax=ax)
ax.set(xlabel='Time',ylabel='Pearson r')
ax.set(title=f"Overall Pearson r = {np.round(overall_pearson_r,2)}")
plt.show()

```
