```python
import tensorflow as tf

from mlxtend.regressor import StackingCVRegressor,StackingRegressor
from mlxtend.data import boston_housing_data
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
# demo3
# 进行了k折交叉验证
x, y = boston_housing_data()
x = x[:100]
y = y[:100]


# 划分数据集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)


# 初始化基模型
lr = LinearRegression()
svr_lin = SVR(kernel='linear', gamma='auto')
ridge = Ridge(random_state=2019,)
lasso =Lasso()


# 加载已保存的模型 model_new.meta
tf.reset_default_graph()
sess = tf.Session()
# saver = tf.train.import_meta_graph('model_new.meta')
saver = tf.train.import_meta_graph("saved_model5/model5.meta")

saver.restore(sess, "saved_model5/model5")

# 创建 TensorFlow Estimator
def model_fn(features, labels, mode):
    # 使用整个已加载的模型作为基本模型
    predictions = tf.identity(sess.run('output_node:0', feed_dict={'input_node:0': features}), name='prediction')
    return tf.estimator.EstimatorSpec(mode, predictions=predictions)

# 用 TensorFlow Estimator 创建模型
tf_estimator = tf.estimator.Estimator(model_fn)

models_yuan = [lr, svr_lin, ridge, lasso]
models = [lr, svr_lin, ridge, lasso, tf_estimator]


print("base model")

# 到这之前是跑通的

# cv=5表示使用5折交叉验证，堆叠回归模型使用了5折交叉验证
# 所以它会在训练数据上训练5次，每次使用不同的验证集
# for model in models:  是基本回归模型的训练
for model in models_yuan:
    score = cross_val_score(model, x_train, y_train, cv=5)
    print(score.mean(), "+/-", score.std())
    
print("------------")

# stacking训练
# sclf = StackingCVRegressor(regressors=models, meta_regressor=lasso)
# 有报错，StackingCVRegressor 预期包含 Scikit-Learn 兼容的估算器，
# 而 TensorFlow Estimator 不是 Scikit-Learn 兼容的估算器，因此会导致错误。
# 替换为Scikit-Learn 的 StackingRegressor 类，
# 它允许包含不同的 Scikit-Learn 估算器和/或其他 Scikit-Learn 兼容的估算器

sclf = StackingRegressor(regressors=models, meta_regressor=lasso)
# 还是不兼容



# 训练回归器
print("stacking model")
score = cross_val_score(sclf, x_train, y_train, cv=5)
print(score.mean(), "+/-", score.std())

sclf.fit(x_train, y_train)
pred = sclf.predict(x_test)
print("loss is {}".format(mean_squared_error(y_test, pred)))
```

```python
# 获取训练数据功能：加载 LSTM 的训练数据
# 输入: batch_size, time_step, train_begin, train_end
# 输出: batch_index, train_x, train_y

def get_train_data(batch_size, time_step, train_begin, train_end):
    batch_index = []
    
     
    
    # 规范化数据
    scaler_for_x = MinMaxScaler(feature_range=(0, 1))
    scaled_x_data = scaler_for_x.fit_transform(data)
    
    scaler_for_y = MinMaxScaler(feature_range=(0, 1))
    scaled_y_data = scaler_for_y.fit_transform(data[:, 0].reshape(-1, 1))
    

    # 获得训练数据
    normalized_train_data = scaled_x_data[train_begin:train_end]
    train_x, train_y = [], []
    for i in range(len(normalized_train_data) - time_step):
        if i % batch_size == 0:
            batch_index.append(i)
        x = normalized_train_data[i:i + time_step, 1:6]
        y = normalized_train_data[i+1:i + time_step+1, 0, np.newaxis]
        train_x.append(x.tolist())
        train_y.append(y.tolist())
    
    batch_index.append((len(normalized_train_data) - time_step))
    return batch_index, train_x, train_y 
```

```python
import tensorflow as tf
from mlxtend.regressor import StackingCVRegressor
from mlxtend.data import boston_housing_data
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
from sklearn.base import BaseEstimator, RegressorMixin

# 加载波士顿房价数据集
x, y = boston_housing_data()
x = x[:100]
y = y[:100]

# 划分数据集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# 创建一个包装 TensorFlow Estimator 的 Scikit-Learn 估算器
class TensorFlowEstimator(BaseEstimator, RegressorMixin):
    def __init__(self, model_path):
        self.model_path = model_path
        self.model = None

    def fit(self, X, y):
        self.model = tf.saved_model.load(self.model_path)

    def predict(self, X):
        predictions = self.model(X)
        return predictions['output'].numpy()
    

    
# 指定你的 TensorFlow 模型路径
tf_model_path = "saved_model"  # 替换为你的 TensorFlow 模型路径
    
# 加载SavedModel
with tf.Session() as sess:
    tf.saved_model.loader.load(sess, [tf.saved_model.tag_constants.SERVING], tf_model_path)
    loaded_model = tf.get_default_graph()
    
    

# 创建 TensorFlow Estimator 并包装模型
tf_estimator = TensorFlowEstimator(model_path=tf_model_path)  # 替换为你的 TensorFlow Estimator



# 初始化基本回归模型
lr = LinearRegression()
svr_lin = SVR(kernel='linear', gamma='auto')
ridge = Ridge(random_state=2019)
lasso = Lasso()


# 添加估算器对象到基本模型列表
models = [lr, svr_lin, ridge, lasso, tf_estimator]

print("base model")
for model in models:
    score = cross_val_score(model, x_train, y_train, cv=5, scoring='neg_mean_squared_error')
    print(-score.mean(), "+/-", score.std())

# 初始化堆叠回归模型
sclf = StackingCVRegressor(regressors=models, meta_regressor=lasso)

# 训练堆叠模型
print("stacking model")
score = cross_val_score(sclf, x_train, y_train, cv=5, scoring='neg_mean_squared_error')
print(-score.mean(), "+/-", score.std())

sclf.fit(x_train, y_train)
pred = sclf.predict(x_test)
print("loss is {}".format(mean_squared_error(y_test, pred)))
```
