```python
import pandas as pd

# 创建一个时间范围，从2019年10月1日00:00开始，到2020年10月31日23:59结束，时间间隔为15分钟
date_range = pd.date_range(start="2019-10-01 00:00", end="2020-10-31 23:59", freq="15T")

# 创建一个包含时间戳的DataFrame
df = pd.DataFrame({'timestamp': date_range})

# 现在，你可以读取你的CSV文件并将新的时间戳列与其合并
# 假设你的CSV文件名为data.csv，包含其他数据列
data = pd.read_csv('123123.csv')

# 合并DataFrame，假设你要将新的时间戳列添加到已有的数据中
merged_df = pd.concat([data, df], axis=1)

# 如果需要，你可以将合并后的DataFrame保存回CSV文件
# merged_df.to_csv('data_with_timestamp.csv', index=False)
```
